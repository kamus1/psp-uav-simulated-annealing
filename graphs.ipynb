```python
import matplotlib.pyplot as plt
import numpy as np


n_rows, n_cols = 8, 8
num_visitas = 4 

np.random.seed(7282025) 
puntos = set()
while len(puntos) < num_visitas:
    puntos.add((np.random.randint(0, n_rows), np.random.randint(0, n_cols)))
puntos = list(puntos)


base = (0, 0)


ruta = [base] + puntos


fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(0, n_cols)
ax.set_ylim(0, n_rows)
ax.set_xticks(range(n_cols))
ax.set_yticks(range(n_rows))
ax.grid(True, which='both', color='lightgray', linestyle='-', linewidth=1)
ax.invert_yaxis()

ax.tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)

for i, (r, c) in enumerate(ruta):
    cx, cy = c + 0.5, r + 0.5 
    if (r, c) == base:
        ax.scatter(cx, cy, color='blue', s=130, label='Base', zorder=3)
    else:
        ax.scatter(cx, cy, color='red', s=130, label='Punto de interÃ©s' if i == 1 else "", zorder=3)
        ax.text(cx + 0.25, cy - 0.25, f"C{i}", fontsize=10, color='black')


for i in range(len(ruta) - 1):
    r1, c1 = ruta[i]
    r2, c2 = ruta[i + 1]
    ax.annotate("",
                xy=(c2 + 0.5, r2 + 0.5),
                xytext=(c1 + 0.5, r1 + 0.5),
                arrowprops=dict(arrowstyle="->", color='green', lw=2),
                zorder=2)

ax.set_title("", fontsize=13)
ax.legend(loc='upper right')
plt.show()
```
